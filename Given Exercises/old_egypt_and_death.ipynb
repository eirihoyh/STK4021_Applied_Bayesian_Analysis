# Egyptian Data - Death of Men and Women in Egpyt for 2100 year ago

In this task we will be using a Weibull distribution:

$$F(y) = 1 - e^{-(\frac{y}{a})^b}$$ 

$$\rightarrow f(y) = e^{-(\frac{y}{a})^b} \cdot \frac{b y^{b-1}}{a^b}$$

Here, y represent either male or female and _a_ and _b_ are what we need to find to maximize our function. The way we are going to maximize our function is by minimizing the negative log likelihood function. By minimizing the negative log likelihood, we'll maximize our "hyperparameters" _a_ and _b_. Our log ML will have this formula:

$$\sum_{i=1}^n -\left(\frac{y_i}{a}\right)^b + log(b) + (b-1) \cdot log(y_i) - b \cdot log(a)$$

To do this manually, we'll have to first derivate, with respect to a, equal to zero, then do the same thing with b. We also want to get the inverse maximum information matrix (aka Hessian), but that is extremy hard to solve manually. So, what we are going to do, is to use Python and R-studio to solve this problem for us.

First, lets read in the data and look at how it is distributed:

```python
with open("../data/egypt_data", "r") as f:
    describe_data = [f.readline() for _ in range(5)]
    data = f.readlines(-1)
```

```python
describe_data
```

```python
# from string to float
import numpy as np
men = np.array([float(elem.replace("\n","")) for elem in data[:82]])
women = [float(elem.replace("\n","")) for elem in data[82:]]

n_men = len(men)
n_women = len(women)
```

Let's look at how long each gender suvived:

```python
import matplotlib.pyplot as plt

y_val = np.arange(0, max(max(men), max(women)))
surv_men = 0*y_val
surv_women = 0*y_val

for i in y_val:
    i = int(i)
    surv_men[i] = np.mean(1*(men >= y_val[i]))  # % of men who is x or older (e.g. 40 and older)
    surv_women[i] = np.mean(1*(women >= y_val[i]))
    
plt.plot(y_val, surv_men, 'r--', label="men")
plt.plot(y_val, surv_women, 'b-', label="women")
plt.xlabel("Age")
plt.ylabel("survival curves")
plt.legend()
plt.show()
```

From the plot above, it seems like there is more men alive that are 22 and older, than there is women (aka more men than women). Let's see if this number is significant.

In this task, we want to find our:

1. $\gamma = Pr(\text{live longer than 40.0 years})$, for both male and female. Is the ratio between the two genders different too?
2. $\mu = median = a\cdot (log(2))^{1/b}$, for both male and female, and do they differ?

Now, we'll construct our log maximum Likelihood functions for both male and female.

First men:

```python
import scipy.optimize as opti

def logL_men(ab):
    a = ab[0]
    b = ab[1]
    
    aux_men = -(men/a)**b + np.log(b) + (b-1)*np.log(men) - b*np.log(a)
    return sum(aux_men)

def minuslogL_men(ab):
    return -logL_men(ab)

optimal_params = opti.minimize(minuslogL_men, [13.,1.1])
x_men = optimal_params["x"]
a_men, b_men = x_men[0], x_men[1]

print(round(a_men, 4), round(b_men, 4))
```

I was not able to get the right Hessian matrix using this optimization method, so had to use RStudio's nlm to find a solution (both Python and RStudio found the same a and b values for men):

```python
import matplotlib.pyplot as plt
image = plt.imread('../data/rstudio_hessian_men_egypt_data.jpg')
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(image)
plt.show()
```

```python
hessian_men = np.array([[0.110718, -0.8131067], 
                    [-0.8131067, 68.8313613]])
covariance_matrix_men = np.linalg.inv(hessian_men)  # aka our covariance matrix
sigma_a_men, sigma_b_men = np.sqrt(np.diag(covariance_matrix_men))

print(f"Maximum estimates men: {round(a_men, 4), round(b_men, 4)}\n")
print(f"Sigma men: {round(sigma_a_men, 4), round(sigma_b_men, 4)}\n")
print(f"Covariance matrix men:\n{covariance_matrix_men}")
```

Now we do the same for women:

```python
def logL_women(ab):
    a = ab[0]
    b = ab[1]
    
    aux_women = -(women/a)**b + np.log(b) + (b-1)*np.log(women) - b*np.log(a)
    return sum(aux_women)

def minuslogL_women(ab):
    return -logL_women(ab)

optimal_params = opti.minimize(minuslogL_women, [13.,1.1])
x_women = optimal_params["x"]
a_women, b_women = x_women[0], x_women[1]

print(round(a_women, 4), round(b_women, 4))
```

Again, I was not able to produce a consistant Hessian matrix using Python, so had to use RStudio:

```python
import matplotlib.pyplot as plt
image = plt.imread('../data/rstudio_hessian_women_egypt_data.jpg')
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(image)
plt.show()
```

```python
hessian_women = np.array([[0.1697612, -0.8857788], 
                         [-0.8857788, 47.3575108]])
covariance_matrix_women = np.linalg.inv(hessian_women)  # aka our covariance matrix
sigma_a_women, sigma_b_women = np.sqrt(np.diag(covariance_matrix_women))

print(f"Maximum estimates women: {round(a_women, 4), round(b_women, 4)}\n")
print(f"Sigma women: {round(sigma_a_women, 4), round(sigma_b_women, 4)}\n")
print(f"Covariance matrix women:\n{covariance_matrix_women}")
```

Now that we have got all we need, we can do our "lazy baysian" analysis. 

First, we will tackle our first problem, what's the probability that men/women live longer than 40 years?

```python
import scipy.stats as stats

y0 = 40 # 40 years
sim = 10**5 # number of itterations/numbers/simulations

mean_men = np.array([a_men, b_men])

theta_men = stats.multivariate_normal.rvs(mean=mean_men, cov=covariance_matrix_men, size=sim)
plt.plot(theta_men[:,0], theta_men[:,1], 'ro')
plt.xlabel("theta a_men")
plt.ylabel("theta b_men")
plt.show()
```

```python
mean_women = np.array([a_women, b_women])

theta_women = stats.multivariate_normal.rvs(mean=mean_women, cov=covariance_matrix_women, size=sim)
plt.plot(theta_women[:,0], theta_women[:,1], 'ro')
plt.xlabel("theta a_women")
plt.ylabel("theta b_women")
plt.show()
```

Now, to calculate 

$$\gamma = Pr(Y_i \geq 40)$$

We can use our Weibull distribution:

$$\gamma = Pr(Y_i \geq 40) = 1 - Pr(Y_i \leq 40) = 1 - (1 - e^{-(\frac{y}{a})^b}) = e^{-(\frac{y}{a})^b}$$

```python
prob_men = np.exp(-(y0 / theta_men[:,0])**theta_men[:,1])
prob_women = np.exp(-(y0 / theta_women[:,0])**theta_women[:,1])

print(f"5%, 50% and 95% quantile for men being older than 40:\n{np.quantile(prob_men, [0.05, 0.5, 0.95])}\n")
print(f"5%, 50% and 95% quantile for women being older than 40:\n{np.quantile(prob_women, [0.05, 0.5, 0.95])}\n")
```

It seems like it is much more probable to live longer if you are male (5% quantile for men is the same as the 95% quantile for women).

Now, let's look at the median:

The median is an equal calcualtion to derive the formula:

$$F(y) = 0.5$$

$$\rightarrow e^{-(\frac{y}{a})^b} = 0.5 \rightarrow y = a\cdot log(2)^{1/b)}$$

```python
median_men = theta_men[:,0]*(np.log(2))**(1/theta_men[:,1])
median_women = theta_women[:,0]*(np.log(2))**(1/theta_women[:,1])

plt.plot(median_men, median_women, 'bo')
plt.show()
```

```python
diff = median_men - median_women  # difference between median women and men

plt.hist(diff, bins=70)
plt.show()
```

```python
print(f"5%, 50% and 95% quantile for median men:\n{np.quantile(median_men, [0.05, 0.5, 0.95])}\n")
print(f"5%, 50% and 95% quantile for median women:\n{np.quantile(median_women, [0.05, 0.5, 0.95])}\n")
print(f"5%, 50% and 95% quantile for median difference:\n{np.quantile(diff, [0.05, 0.5, 0.95])}\n")
```

So, if we are pickey statisticians, we would say that we can not say that there is a significant difference between men and women since our difference median quantiles reaches both values from over and under zero. But, that is just slightly. It seems like there is a difference (if we had only looked at 10% and 90% quantiles, we would have said that there is a difference).

So, what we can say in this problem is that there seems like there is a difference between men and women (men lives longer than women) but we can not be 100% sure without.
```python
#------old part from above------------
y_val = np.arange(0, max(max(men), max(women)))
surv_men = 0*y_val
surv_women = 0*y_val

for i in y_val:
    i = int(i)
    surv_men[i] = np.mean(1*(men >= y_val[i]))  # % of men who is x or older (e.g. 40 and older)
    surv_women[i] = np.mean(1*(women >= y_val[i]))
#---------------------------------------

# Weibull and our estimated values:

surv_men_weibull = np.exp(-(y_val / a_men)**b_men)
surv_women_weibull = np.exp(-(y_val / a_women)**b_women)
    
plt.figure(figsize=(15,10))
plt.plot(y_val, surv_men, 'r-', label="surv men")
plt.plot(y_val, surv_men_weibull, 'r--', label="surv men weibull")
plt.plot(y_val, surv_women, 'b-', label="surv women")
plt.plot(y_val, surv_women_weibull, 'b--', label="surv women weibull")
plt.xlabel("Age")
plt.ylabel("survival curves")
plt.legend()
plt.show()
```
