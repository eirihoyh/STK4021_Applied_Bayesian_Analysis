# San Francisco - N cable cars?

In San Francisco there exist N calbe cars, numbered sequentially from 1 to N. We see one random cable car thats numbered 203. We want to estimate _N_.

## (a)

Assume that our prior belifes on _N_ is a gemoetric distribution with mean 100:

$$p(N) = \frac{1}{100} \frac{99}{100}^{N-1}, \quad N=1,2,...$$

What is the posterior distribution for _N_?

To be able to caluclate a posteiror distribution, we need a model/likelihood. In this problem we'll use a uniform distribution:

$$f(y|N) = \frac1N$$

Thus the posterior distribution will be:

\begin{align*}
\pi(N|y) &= \frac{1}{f_{marg}(y)}  \frac{1}{100} \frac{99}{100}^{N-1} \frac1N \\
&\propto  \frac{99}{100}^{N-1} \frac1N, \quad \text{for }N \geq 203
\end{align*}

Note that we need to find our constants in this problem, but we can simulate it!y

## (b) 

What is the posterior mean and standard deviation of _N_?

This one can be a be a bit tricky to by hand, but easy to do using computer power. We know that the posterior distribution is 

$$\pi(N|y) = \frac1c \frac{99}{100}^{N-1} \frac1N$$

Since our posterior is a pmf, we know that it should sum to 1. Let's find our constant:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

obs = 203
N_val = np.arange(1, 10**5, 1)

prior = 1/100 * (99/100)**(N_val-1)  # just for visualization later
pre_post = (99/100)**(N_val-1) * (1/N_val) * (obs<=N_val) 
c = sum(pre_post)
print(f"Constant: {c}")
print(f"1/c = {c**-1}")
```

Thus our posterior will then be:

```python
post = 1/c * pre_post
```

Let's visualize the prior and posterior distribution:

```python
plt.plot(N_val[1:500], prior[1:500], label="prior")
plt.plot(N_val[1:500],post[1:500], label="post")
plt.legend()
plt.show()
```

Now, for the mean and stdandard deviations for the posterior distribution. Formula for the expected value:

\begin{align*}
E(N) &= \sum_{N=203}^{\infty} N \cdot \frac1c \frac{99}{100}^{N-1} \frac1N \\
&= \frac1c \sum_{N=203}^{\infty} \frac{99}{100}^{N-1}
\end{align*}

For the standard deviation we get the following formula:

\begin{align*}
SD(N) &= \sqrt{Var(N)} \\
&= \sqrt{E(N^2) - (E(N))^2} \\
&= \sqrt{\frac1c \sum_{N=203}^{\infty} N \cdot \frac{99}{100}^{N-1} - (E(N))^2}
\end{align*}

Lets find them by using Python:

```python
mean_post = 1/c * sum([(99/100)**(N-1) for N in range(obs, 10**5, 1)])
sd_post = np.sqrt(1/c *sum([N*(99/100)**(N-1) for N in range(obs, 10**5, 1)]) - mean_post**2)

print(f"Mean: {round(mean_post, 3)}\nSD: {round(sd_post, 3)}")
```

__Extra from Nils__

Say now that we now have 3 observations instead of 1. Our observations are 203, 157 and 222. Now, our likelihood would look like this:

$$f(y_1, y_2, y_3|N) = \left(\frac{1}{N}\right)^3$$

and our posterior will then become:

$$\pi(N|y_1,y_2,y_3) = \frac1c \cdot \left(\frac{1}{N}\right)^3 \frac{99}{100}^{N-1}$$

Calculate mean and standard deviation using the "new" posterior and compare the results.

```python
obs3 = [203, 157, 222]
pre_post3 = (99/100)**(N_val-1) * (1/N_val)**3 * (max(obs3)<=N_val) 
c = sum(pre_post3)
post3 = 1/c * pre_post3

mean_post3 = 1/c * sum([(1/N)**2 * (99/100)**(N-1) for N in range(max(obs3), 10**5, 1)])
sd_post3 = np.sqrt(1/c *sum([1/N * (99/100)**(N-1) for N in range(max(obs3), 10**5, 1)]) - mean_post3**2)

print(f"Mean: {round(mean_post3, 3)}\nSD: {round(sd_post3, 3)}")
```

```python
plt.plot(N_val[1:500], prior[1:500], label="prior")
plt.plot(N_val[1:500],post[1:500], label="old post")
plt.plot(N_val[1:500],post3[1:500], label="new post")
plt.legend()
plt.show()
```

Let's compare the mean, standard deviation, 0.05 quantile, 0.50 quantile and 0.95 quantile between the prior, the posterior of just one observation and the posterior with three observations:

```python
mean_prior = sum(N_val * prior)  # mean of prior
sigma_prior = np.sqrt(sum((N_val - mean_prior)**2 * prior))  # standard deviation for prior

cum_prior = np.cumsum(prior)
cum_post1 = np.cumsum(post)
cum_post3 = np.cumsum(post3)

low_prior, mid_prior, high_prior = min(N_val[cum_prior >= 0.05]), min(N_val[cum_prior >= 0.5]), min(N_val[cum_prior >= 0.95])
low_post1, mid_post1, high_post1 = min(N_val[cum_post1 >= 0.05]), min(N_val[cum_post1 >= 0.5]), min(N_val[cum_post1 >= 0.95])
low_post3, mid_post3, high_post3 = min(N_val[cum_post3 >= 0.05]), min(N_val[cum_post3 >= 0.5]), min(N_val[cum_post3 >= 0.95])

check_prior = [mean_prior, sigma_prior, low_prior, mid_prior, high_prior]
check_post1 = [mean_post, sd_post, low_post1, mid_post1, high_post1]
check_post3 = [mean_post3, sd_post3, low_post1, mid_post3, high_post3]

print("mean, std, low, mid, high")
print(check_prior)
print(check_post1)
print(check_post3)
```

## (c)

Choose a reasonable "noninformative" prior.

Can try to use Jeffreys prior:

\begin{align*}
L(N) &= \frac1N \\
\rightarrow l(N) &= -log(N) \\
\rightarrow \frac{d}{dN} l(N) &= -\frac1N \\
\rightarrow \frac{d^2}{dN^2} &= \frac{1}{N^2} \\
\rightarrow I(N) &= -E\left(\frac{1}{N^2} \right) = -\frac{1}{N^2} \\
\rightarrow p_{jeff}(N) &= \left[|-\frac{1}{N^2}|\right]^{1/2} = \frac{1}{N}
\end{align*}

The posterior:

\begin{align*}
\pi(N|y) &= \frac1c \frac1N \cdot \frac1N \\
&\propto \frac{1}{N^2}
\end{align*}

Lets try to do the same as we did in (b):

```python
obs = 203
N_val = np.arange(1, 10**4, 1)

prior = 1/N_val  # just for visualization later

pre_new_post = 1/(N_val**2) * (obs<=N_val) 
c = sum(pre_new_post)
print(f"Constant: {c}")
print(f"1/c3 = {c**-1}")
```

```python
new_post = 1/c * pre_new_post
```

```python
plt.plot(N_val[1:1000], post[1:1000], label="prev post")
plt.plot(N_val[1:1000], new_post[1:1000], label="new post")
plt.legend()
plt.show()
```

As we can see from the plot, the "new post" is lower and deacreases less than our "prev post". Our "prev post" is much steeper and therefore we can state that it is more confident of its answer than "new post". We can also look at the mean and std to verify: 

```python
mean_new_post = sum(N_val*new_post)
sd_new_post = np.sqrt(sum((N_val - mean_new_post)**2 * new_post))
print(f"Old post:\nMean: {round(mean_post, 3)}, std: {round(sd_post, 3)}")
print(f"New post:\nMean: {round(mean_new_post, 3)}, std: {round(sd_new_post, 3)}")
```

The mean and std for our new post is much larger for the "new post" than for the "prev post". Meaning, again, that the old posterior is more confident of its answer than the new posterior.
