_We are given that_

* 674 given control --> 39 died
* 680 receiving the treatment --> 22 died

_Assume that the outcumes are iid binomially distributed, with porb of death $p_0$ and $p_1$ under the control and treatmen, respectivly.

## (a)

_Set up a noninformative prior distribution on ($p_0,p_1$) and obtain posterior simulations._

A noninformative prior can be found by calculating Jeffrey's prior. The general formula is 

$$\pi_{jeff}(\theta) = \left|I(\theta)\right|^{1/2} $$

where $I(\theta)$ is the Fisher information. Since we know that our likelihood is a binomial, we can try to calulate the derivative and second dervative of the log-likelihood function:

\begin{align*}
l(p) &= log(L(p)) \\
&= log\binom{n}{y} + y \cdot log(p) + (n-y)log(1-p) \\
\frac{d}{dp}l(p) &= \frac{y}{p} - \frac{n-y}{1-p} = \frac{y-np}{p(1-p)}
\end{align*}

It is easier to deal with the "single" derivative of the log-likelihood function. We can now calculate the Fisher information:

\begin{align*}
I(p) &= Var(\frac{y-np}{p(1-p)} \\
&= \frac{1}{p^2(1-p)^2} Var(y) \\
&= \frac{np(1-p)}{p^2(1-p)^2} \\
&= \frac{n}{p(1-p)} \\
\end{align*}

thus we get that Jeffreys prior is

$$\pi_{jeff}(p) = |I(p)|^{1/2} \propto \frac{1}{\sqrt{p(1-p)}}$$

which corresponds to $Beta(\frac12, \frac12)$. To illustrate that they are the same, we can visualize:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
vals = np.arange(0,1.00, 0.0001)
plt.plot(stats.beta.pdf(vals ,0.5, 0.5))
plt.show()
```

```python
jeff = [1/(np.sqrt(p0*(1-p0))) for p0 in np.arange(0.0001,0.999999,0.0001)]
plt.plot(jeff)
plt.show()
```

We can see that they resmple the same distribution. 

Now for the posterior distribution, we get that

$$p_i|data \sim Beta(\frac12 + y_i. \frac12 + n_i - y_i, \quad i=0,1$$


We have 
* $n_0 = 674, y_0 = 39$ (controll group)
* $n_1 = 680, y_1 = 22$ (treatment group)

Let's plot the posterior for both groups:

```python
n0, y0 = 674, 39  # n partisipant, y deaths
n1, y1 = 680, 22

p_vals = np.arange(0, 0.2, 0.001)  # reduced potensial p_vals to 0.2, there where the "action" is

a_prior, b_prior = 1/2, 1/2

a0_post, b0_post = a_prior+y0, b_prior+n0-y0
a1_post, b1_post = a_prior+y1, b_prior+n1-y1


f0_data = stats.beta.pdf(p_vals, a0_post, b0_post)  # controll
f1_data = stats.beta.pdf(p_vals, a1_post, b1_post)  # treatment


plt.plot(p_vals, f0_data, label="controll group")
plt.plot(p_vals, f1_data, label="Treatment group")
plt.legend()
plt.show()
```

## (b)

_Use the odds ratio $\frac{p_1/(1-p_1)}{p_0/(1-p_0)}$ to sumarize the posterior_

```python
sims = 10**5

p0_sim = stats.beta.rvs(a0_post,b0_post,size=sims)
p1_sim = stats.beta.rvs(a1_post, b1_post, size=sims)

odds_ratio = (p1_sim/(1-p1_sim))/(p0_sim/(1-p0_sim))

plt.hist(odds_ratio, bins=40)
plt.show()
```

Can also obtain the credibility interval:

```python
print(f"95% credibility interval: {np.quantile(odds_ratio, [0.025, 0.5, 0.975])}")
```

The plot and credibility interval estimates that the odds is higher for $p_0$ than what it is for $p_1$. This indicates that the treatment actually works bettwe than controll since lower odds corresponds to lower probability of death.

## (c)

_Discuss the sensitivity of your inference to your choice of noninformative prior density._

The weakness of a noninformative prior is that we are very open to accept that the data given to us is the "ground thruth". Iow, it is easier to convince our model that the data we have observed is right. So, for example, if there is actually no difference between controll and treatment, but we see some more deaths in controll than treatment (which is by chance) our model will favour the treatment falsely.
